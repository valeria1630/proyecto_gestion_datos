## Acondicionamiento de los datos

A continuación se verificará la condición numérica de las variables que sean números, con el propósito de transformar dichos valores para que posteriormente puedan ser procesados.

```python
columnas_verificar = ['Año', 'Ha_sembradas', 'Ha_cosechadas', 'Tiempo_produccion', 'Rendimiento']

for columna in columnas_verificar:
    # Verificar si los valores de la columna son numéricos
    es_numerico = []
    for valor in datos[columna]:
        try:
            pd.to_numeric(valor)
            es_numerico.append(True)
        except ValueError:
            es_numerico.append(False)

    # Contar cuántos son numéricos y cuántos no
    conteo_numericos = sum(es_numerico)
    conteo_no_numericos = len(es_numerico) - conteo_numericos

    # Mostrar los resultados para cada columna
    print('Cantidad de valores numéricos en', columna, ':', conteo_numericos)
    print('Cantidad de valores no numéricos en', columna, ':', conteo_no_numericos)
    print()  # Separación entre resultados de columnas
```

Con estos valores claros, se procede a transformar las variables a valores numéricos, tal y como se muestra a continuación.

```python
columnas_a_transformar = ['Año', 'Ha_sembradas', 'Ha_cosechadas', 'Tiempo_produccion']

# Reemplazar comas y puntos y luego convertir a numérico para cada columna
for columna in columnas_a_transformar:
    datos[columna] = datos[columna].str.replace(',', '').str.replace('.', '').astype(float)

# Verifica los cambios
print(datos.dtypes)
```

Ahora bien, para representar de una mejor manera el periodo, lo que se hará es eliminar el año del texto, es decir que si el periodo representaba el año 2006B, ahora el periodo será únicamente B, esto con el propósito de facilitar el procesamiento y análisis del dataset. Es por eso, que antes de la modificación de esta columna, se realiza un análisis exploratorio de la misma.

```python
# Calcular el conteo de apariciones para cada dato en la columna "Periodo"
conteo_apariciones = dataset['Periodo'].value_counts()

# Imprimir el conteo de apariciones
print('Conteo de apariciones para cada dato en la columna "Periodo":')
print(conteo_apariciones)
```

Con esto, se aprecia que hay múltiples tipos de valores dentro de esta columna. Es por esto que se decide reasignar algunas de estos valores, de tal forma que se obtengan tres categorías: A, B y Sin Temporada, con el propósito de evitar la eliminación de datos, al tiempo que mejorando el análisis de los mismos.

```python
# Función para clasificar el periodo
def clasificar_periodo(periodo):
    if periodo[-1] in ['A', 'B']:
        return periodo[-1]
    else:
        return "Sin Temporada"

# Aplicar la clasificación a la columna 'Periodo'
datos['Periodo'] = datos['Periodo'].apply(clasificar_periodo)

# Filtrar solo las clasificaciones que nos interesan
clasificaciones_interesantes = ['A', 'B', 'Sin Temporada']
datos = datos[datos['Periodo'].isin(clasificaciones_interesantes)]

# Imprimir el resultado
conteo_apariciones = datos['Periodo'].value_counts()
print('Conteo de apariciones para cada dato en la columna "Periodo":')
print(conteo_apariciones)
```

## Análisis por Departamento

```python
# Agrupar por departamento y año, contar el número de cultivos
cultivos_por_departamento_y_año = datos.groupby(['Depto', 'Año']).size().unstack().fillna(0)
# Gráfico de barras para la cantidad de cultivos por departamento en cada año
plt.figure(figsize=(12, 6))
sns.heatmap(cultivos_por_departamento_y_año, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Cantidad de cultivos por departamento en cada año')
plt.xlabel('Año')
plt.ylabel('Departamento')
plt.show()
```

En la gráfica previamente mencionada, se presenta un análisis de la cantidad de cultivos por departamento a lo largo de 13 años, desde 2006 hasta 2018. Entre los hallazgos más destacados, se observa que departamentos como Antioquia, Boyacá, Cundinamarca, Valle del Cauca y Huila han experimentado un crecimiento gradual en la cantidad de sus cultivos por año. Este crecimiento se refleja en la gráfica mediante una tendencia constante de aumento a lo largo del período analizado. En conjunto, estos hallazgos sugieren una consolidación y expansión sostenida de la actividad agrícola en estos departamentos durante el período estudiado. Este crecimiento gradual puede atribuirse a una serie de factores, como la adopción de prácticas agrícolas más eficientes, inversiones en infraestructura agrícola, condiciones climáticas favorables y posiblemente políticas gubernamentales que respaldan el desarrollo agrícola en estas regiones. Esto no solo puede contribuir al crecimiento económico regional, sino que también puede tener un impacto significativo en la seguridad alimentaria y la creación de empleo en estas áreas, lo que hace que este análisis sea de gran importancia para la toma de decisiones a nivel local y nacional por parte de la empresa.

```python
# Eliminar valores nulos en la columna de rendimiento
datos_limpios = datos.dropna(subset=['Rendimiento'])

# Crear el boxplot del rendimiento por departamento
plt.figure(figsize=(12, 6))
ax = sns.boxplot(data=datos_limpios, x='Depto', y='Rendimiento', palette='pastel')
ax.set_title('Rendimiento de los Cultivos por Departamento')
ax.set_xlabel('Departamento')
ax.set_ylabel('Rendimiento (t/ha)')
plt.xticks(rotation=45)
plt.show()
```

Los diagramas de cajas, también conocidos como boxplots, desempeñan un papel fundamental en la representación y comprensión de datos, como el rendimiento de cultivos por departamentos. Estas visualizaciones gráficas nos permiten examinar de manera efectiva la variabilidad y la distribución de los datos, así como identificar valores atípicos . Sin embargo, en ocasiones, para un análisis más preciso de la distribución de los datos, es beneficioso realizar un análisis detallado excluyendo los valores atípicos. Al eliminar estos valores inusuales, podemos enfocarnos en la caja central del diagrama, y obtener una visión más clara de la variabilidad dentro de los datos. Esto nos permite identificar tendencias, comparar departamentos y extraer conclusiones significativas sobre el rendimiento de cultivos en cada área, sin que los valores atípicos afecten el estudio general de la distribución de datos. Como se va observar en la siguiente gráfica.

```python
# Eliminar valores nulos en la columna de rendimiento
datos_limpios = datos.dropna(subset=['Rendimiento'])

# Crear el boxplot del rendimiento por departamento con swarm plot
plt.figure(figsize=(12, 6))
ax = sns.boxplot(data=datos_limpios, x='Depto', y='Rendimiento', palette='pastel', showfliers=False)  # No mostrar outliers en el boxplot
ax.set_title('Rendimiento de los Cultivos por Departamento')
ax.set_xlabel('Departamento')
ax.set_ylabel('Rendimiento (t/ha)')
plt.xticks(rotation=45)
plt.show()
```

Si observamos el diagrama de cajas para el rendimiento de los cultivos por departamento, notamos que en el departamento de Antioquia, la mediana es bastante alta, lo que sugiere un rendimiento promedio sólido en comparación con otros departamentos. Además, los bigotes son relativamente cortos, indicando una dispersión limitada de los datos. Sin embargo, notamos algunos valores atípicos por encima del tercer cuartil, lo que podría indicar áreas de cultivo excepcionalmente productivas o circunstancias especiales en ciertos años. Ahora bien, al comparar los departamentos de Antioquia y Boyacá en el diagrama, notamos que ambos tienen medianas relativamente altas, lo que sugiere que, en promedio, han tenido un rendimiento de cultivos similar a lo largo del tiempo. Sin embargo, es importante destacar que la caja que representa Boyacá es considerablemente más grande que la de Antioquia en términos de su amplitud vertical.

Esta diferencia en el tamaño de las cajas nos proporciona información valiosa sobre la variabilidad en el rendimiento de los cultivos en ambos departamentos. En el caso de Antioquia, la caja más compacta indica que su rendimiento de cultivos ha tendido a mantenerse en un rango más estrecho a lo largo de los años, lo que sugiere una mayor consistencia en la producción agrícola. Por otro lado, la caja más amplia de Boyacá señala una mayor variabilidad en su rendimiento de cultivos a lo largo del tiempo.Esta variabilidad podría deberse a diversas razones, como condiciones climáticas cambiantes, fluctuaciones en la inversión agrícola, o posiblemente la diversidad de cultivos cultivados en Boyacá. Es importante tener en cuenta que, si bien Antioquia y Boyacá pueden tener medianas similares en rendimiento de cultivos, la amplitud de la caja de Boyacá sugiere que este departamento ha experimentado una mayor variabilidad en su producción agrícola en comparación con Antioquia.

```python
# Filtrar los 10 departamentos con mayor cantidad de cultivos
top_10_departamentos = datos['Depto'].value_counts().nlargest(10).index
datos_top_10 = datos[datos['Depto'].isin(top_10_departamentos)]

# Agrupar por departamento y ciclo de cultivo y contar la cantidad
conteo_ciclo_por_depto = datos_top_10.groupby(['Depto', 'Ciclo_cultivo']).size().unstack().fillna(0)

# Sumar la cantidad de cultivos por departamento
conteo_cultivos_por_depto = conteo_ciclo_por_depto.sum(axis=1)

# Ordenar los departamentos por la cantidad de cultivos
orden_departamentos = conteo_cultivos_por_depto.sort_values(ascending=False).index
conteo_ciclo_por_depto = conteo_ciclo_por_depto.loc[orden_departamentos]

# Crear el gráfico de barras
conteo_ciclo_por_depto.plot(kind='bar', stacked=False, figsize=(12, 6))
plt.xlabel('Departamento')
plt.ylabel('Cantidad de cultivos')
plt.title('Cantidad de cultivos por ciclo de cultivo en los 10 departamentos con más cultivos')
plt.xticks(rotation=45)
plt.legend(title='Ciclo de cultivo')
plt.show()
```

Ahora bien, se evalúa el número de dieferntes tipos de cultivos que son cultivados en un ciclo especifico en los 10 departamentos que tienen la mayor cantidad de cultivos en un país. Donde se busca identificar la cantidad de cultivo que ha sido madurada con exito, el ciclo del cultivo como el periodo específico durante el cual se lleva la siembra, crecimiento, y cosecha de los cultivos. Con una liosta de los 10 departamentos que tienen mayor reproducción agrícola que se determinaron anteriormente. De esta forma, se evidencia que el departamento de Boyaca tiene el mayor ciclo de cultivo exitoso de manera transitoria. Lo quer indica que el período durante el cual se realiza la transición de un cultivo a otro se considera en este caso un ciclo transitorio. Durante esta fase, es posible que se deban tomar medidas para preparar el suelo, adaptar las prácticas de manejo y ajustar la gestión de plagas y enfermedades. Para la adpatación a los cambvios climaticos y la introducción de nuevas técnicas si son necesarias. En el caso de Antioquia tiene de igual manera la cantidad de cultivo por ciclo de vida de cultivo, lo que indica, que el cultivo llega a ser sembrado exitosamente de manera permanenete en la región tanto como de manera transitoria en las épocas con mayores cambios climáticos. Sin embnargo, se evidencia que el depratamemto de Cauca presenta bajas cantidades de cultivo durante las fases transitoria, permanenete y anuales, lo que deja en evidencia las posibles dificultades medioambientales que enfrentan los hombres en esta región. Donde sobre sale la cantidad transitoria, teniendo presente, los cambios de temperatura que se puedan presentar qye afectan la cantidad de cultivo, entre otros factores.

```python
# Filtrar los 10 departamentos con mayor área cultivada
top_10_departamentos = datos.groupby('Depto')['Ha_cosechadas'].sum().nlargest(10).index
datos_top_10 = datos[datos['Depto'].isin(top_10_departamentos)]

# Agrupar por departamento y ciclo de cultivo y sumar el área cosechada
area_cultivada_por_depto = datos_top_10.groupby(['Depto', 'Ciclo_cultivo'])['Ha_cosechadas'].sum().unstack().fillna(0)

# Sumar el área total cultivada por departamento
area_total_por_depto = area_cultivada_por_depto.sum(axis=1)

# Ordenar los departamentos por el área total cultivada
orden_departamentos = area_total_por_depto.sort_values(ascending=False).index
area_cultivada_por_depto = area_cultivada_por_depto.loc[orden_departamentos]

# Crear el gráfico de barras
area_cultivada_por_depto.plot(kind='bar', stacked=False, figsize=(12, 6))
plt.xlabel('Departamento')
plt.ylabel('Área cultivada (Ha)')
plt.title('Área cultivada por ciclo de cultivo en los 10 departamentos con más área cultivada')
plt.xticks(rotation=45)
plt.legend(title='Ciclo de cultivo')
plt.show()
```

Lo que permite evidenciar la cantidad total de tierra empleada para la agricultura en estos departamentos. Identificando las regiones con mayior dedicación de tierra a la agricultura y con mayor potencial para la producción agrícola a gran escala. Igualmente, es importante determinar los cultivos con su ciclo de vida. Ya que su ciclo de crecimiento completo, desde la siembra hasta la cosecha, tiende a ser menod de un año ya que se requieren para cosechas rápidas, y lograr obtener tierra que pueda ser preparada nuevamente para un siguiente cultivo. Por lo tanto su siembra y cosecha solo ocurren en una estación o temporada específica. En el caso del platano, café y cacao, son cultivos que sobresalen por su fuerte inclinación a ser permanenetes. Lo que hace referencia a cultivos con una vida más larga y que pueden producir cosechas durante varios años como lo son estos casos. Ya que son cultivos que en el país con establecidos una vez para proporcionar cosechas durante varias temporadas con un mantenimiento continuo para su cosecha de manera repetida duarnte varios años.

```python
# Obtener los top N cultivos más comunes
N = 15
top_N_cultivos = datos['Cultivo'].value_counts().nlargest(N).index
datos_top_N = datos[datos['Cultivo'].isin(top_N_cultivos)]

# Filtrar solo las columnas de interés
datos_top_N = datos_top_N[['Cultivo', 'Ha_sembradas', 'Ha_cosechadas']]

# Agrupar por cultivo y sumar las hectáreas sembradas y cosechadas
datos_agrupados = datos_top_N.groupby('Cultivo').sum()

# Ordenar los datos por hectáreas sembradas en orden descendente
datos_agrupados = datos_agrupados.sort_values(by='Ha_sembradas', ascending=False)

# Crear el gráfico de barras agrupadas
plt.figure(figsize=(14, 8))
datos_agrupados.plot(kind='bar')
plt.xlabel('Cultivo')
plt.ylabel('Hectáreas')
plt.title('Comparación de Hectáreas Sembradas y Cosechadas para los Top {} Cultivos (Ordenado por Hectáreas Sembradas)'.format(N))
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Tipo de hectárea', labels=['Hectáreas Sembradas', 'Hectáreas Cosechadas'], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Asimismo, es posible vincularlo a la gestión de producción en áreas destinadas a la recuperación para una mejor calidad y rendimiento del fruto, optando por no cosechar todas las áreas sembradas para garantizar mejores cultivos. Por otro lado, frutos como el café que evidencia la mayor diferencia de frecuencias puede ser parte de cosechas selectivas, donde, solo se cosechan productos maduros y de alta calidad, dejando de lado productos no maduros en la planta. Finalmente, las deciciones economicas pueden influir en estas diferencias en conjunto con las condiciones climaticas cambiantes y las diferentes estaciones del año que afectan la siembra y cosecha de los frutos.

En este contexto, es fundamental diferenciar entre cultivos permanentes y cultivos transitorios, ya que estos presentan características y requerimientos distintos.Por ejemplos, departamentos como Antioquia y Valle del Cauca, se destacan por sus ciclos de cultivos permanentes, mostrando una estrategia agrícola que se centra en plantaciones a largo plazo. Estos cultivos, que incluyen frutales y otros, requieren una inversión inicial significativa y un mantenimiento constante.Por otro lado, departamentos como Tolima, que se enfocan en cultivos transitorios, enfrentan una dinámica diferente. Los cultivos transitorios, como cereales y hortalizas, se caracterizan por su ciclo de vida más corto y una rotación más frecuente de cultivos en un mismo terreno. De esta manera, el análisis de datos en estas regiones puede ayudar a determinar cuáles son los cultivos más rentables en cada temporada, planificar la siembra y la cosecha de acuerdo con las condiciones climáticas y de mercado cambiantes, y minimizar los riesgos asociados con la variabilidad en los precios y la demanda.

Asimismo, la observación de las tendencias de cultivo en un período de 13 años (2006-2018) es de gran importancie e influencia en las estrategias de Serviagro, como una empresa agrícola que busca agregar valor a sus productos y liderar en el mercado.Por lo tanto, la constante predominancia de los cultivos de tubérculos y plátanos, combinados con frutales, a lo largo de estos años, es un indicativo de la estabilidad y resistencia de estos cultivos en la región. EEsto resulta valioso para Serviagro, ya que sugiere que estos cultivos son confiables y pueden proporcionar un flujo constante de ingresos. Además, su demanda aparentemente constante en el mercado puede brindar cierta seguridad económica a la empresa a lo largo del tiempo.

Sin embargo, la variabilidad en los periodos A y B, donde se destacan las hortalizas y los cereales, respectivamente, sugiere que existe un mercado en constante cambio que busca una variedad de productos agrícolas. Esta variabilidad ofrece a Serviagro una oportunidad interesante para la diversificación de su producción. Diversificar su oferta agrícola, centrándose en la producción de hortalizas de alta demanda y cereales de alta calidad, podría permitir a Serviagro atender las necesidades cambiantes del mercado y mejorar su competitividad.

De esta manera, el análisis de datos por departamento y por período es esencial para Serviagro en su búsqueda de agregar valor a sus productos y liderar en el mercado agrícola. Este análisis proporciona información clave sobre las zonas más adecuadas para la producción, los cultivos óptimos y la época del año ideal para maximizar las cosechas. Por lo tanto, la empresa puede aprovechar esta información para tomar decisiones informadas y desarrollar prácticas agrícolas que impulsen su crecimiento en el negocio agrícola.
